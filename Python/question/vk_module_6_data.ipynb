# module_6_задача 3
Оцените среднее время ответа веб- сервера с расчётом доверительного интервала.

Предположим у нас есть логи времени ответа веб-сервера на запросы, и мы хотим оценить среднее время ответа этого веб-сервера с учётом неопределенности.
 Для этого мы можем рассчитать доверительный интервал для среднего времени ответа на основе выборочных данных.
 Выборка с временем ответа веб-сервера:

response-times = [150, 130, 140, 160, 180, 170, 190, 200, 175, 165]

Чему равна левая граница доверительного интервала при уровне доверия 95% с точностью до двух знаков после запятой?

# Решние
Для расчета доверительного интервала среднего времени ответа веб-сервера необходимо знать среднее значение выборки, стандартное отклонение и размер выборки.

Среднее значение выборки можно найти, просто посчитав среднее арифметическое:
mean = (150 + 130 + 140 + 160 + 180 + 170 + 190 + 200 + 175 + 165) / 10 = 167

Стандартное отклонение можно найти по формуле:
std = sqrt(sum((x - mean)^2) / (n - 1))
где x - значение из выборки, mean - среднее значение выборки, n - размер выборки.

std = sqrt((150-167)^2 + (130-167)^2 + (140-167)^2 + (160-167)^2 + (180-167)^2 + (170-167)^2 + (190-167)^2 + (200-167)^2 + (175-167)^2 + (165-167)^2 / (10-1)) = 20.21

Размер выборки равен 10.

Для расчета доверительного интервала используем формулу:
CI = mean ± t*(std/sqrt(n))
где CI - доверительный интервал, mean - среднее значение выборки, t - коэффициент Стьюдента для заданного уровня доверия и размера выборки, std - стандартное отклонение, n - размер выборки.

Для уровня доверия 95% и размера выборки 10 коэффициент Стьюдента равен 2.262.

CI = 167 ± 2.262*(20.21/sqrt(10)) = (146.44, 187.56)

Таким образом, левая граница доверительного интервала при уровне доверия 95% с точностью до двух знаков после запятой равна 146.44.

Ответ: 146.44

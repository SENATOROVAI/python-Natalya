# module_5_Задача_4:

 Сравнение производительности двух различных алгоритмов сортировки для массивов большого размера.

 Предположим, у нас есть два различных алгоритма сортировки (Алгоритм А и Алгоритм В), и нам интересно определить, какой из них работает более эффективно для больших массивов данных.

Визборки с временем работы алгоритмов:

algorithm_a_times [3.2, 5.1, 4.3, 6.7, 7.9, 4.8, 5.6, 6.2, 5.9);

algorithm b times 12.9, 4.5, 6.1, 5.8, 4.2, 5.3, 4.7, 6.5, 6.8].

Совет: Для сравнения производительности алгоритмов примените двусторонний непараметрический критерий Манна Уитни.

Чему равно значение полученной статистики с точностью до одного знака после запятой?

```python
from scipy.stats import mannwhitneyu

algorithm_a_times = [3.2, 5.1, 4.3, 6.7, 7.9, 4.8, 5.6, 6.2, 5.9]
algorithm_b_times = [2.9, 4.5, 6.1, 5.8, 4.2, 5.3, 4.7, 6.5, 6.8]

# Применение критерия Манна-Уитни
statistic, p_value = mannwhitneyu(algorithm_a_times, algorithm_b_times, alternative='two-sided')

# Вывод значения статистики U с точностью до одного знака после запятой
print(f'{statistic:.1f}')
```

```python
from scipy.stats import mannwhitneyu

algorithm_a_times = [3.2, 5.1, 4.3, 6.7, 7.9, 4.8, 5.6, 6.2, 5.9]
algorithm_b_times = [2.9, 4.5, 6.1, 5.8, 4.2, 5.3, 4.7, 6.5, 6.8]

# Применение критерия Манна-Уитни
statistic, p_value = mannwhitneyu(algorithm_a_times, algorithm_b_times, alternative='two-sided')

# Вывод значения p-значения с точностью до двух знаков после запятой
print(f'{p_value:.2f}')
```

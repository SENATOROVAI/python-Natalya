```python
# and ‚Äî –ª–æ–≥–∏—á–µ—Å–∫–æ–µ —É–º–Ω–æ–∂–µ–Ω–∏–µ;
# or ‚Äî –ª–æ–≥–∏—á–µ—Å–∫–æ–µ —Å–ª–æ–∂–µ–Ω–∏–µ;
# not ‚Äî –ª–æ–≥–∏—á–µ—Å–∫–æ–µ –æ—Ç—Ä–∏—Ü–∞–Ω–∏–µ.
```

```python
age = int(input('–°–∫–æ–ª—å–∫–æ –≤–∞–º –ª–µ—Ç?: '))
grade = int(input('–í –∫–∞–∫–æ–º –∫–ª–∞—Å—Å–µ –≤—ã —É—á–∏—Ç–µ—Å—å?: '))
if age >= 12 and grade >= 7:
    print('–î–æ—Å—Ç—É–ø —Ä–∞–∑—Ä–µ—à–µ–Ω.')
else:
    print('–î–æ—Å—Ç—É–ø –∑–∞–ø—Ä–µ—â–µ–Ω.')
```

```python
# –±–ª–æ–∫ –∫–æ–¥–∞ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ –ø—Ä–∏ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏–∏ –æ–±–æ–∏—Ö —É—Å–ª–æ–≤–∏–π –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ!
```

```python
age = int(input('–°–∫–æ–ª—å–∫–æ –≤–∞–º –ª–µ—Ç?: '))
grade = int(input('–í –∫–∞–∫–æ–º –∫–ª–∞—Å—Å–µ –≤—ã —É—á–∏—Ç–µ—Å—å?: '))
city = input('–í –∫–∞–∫–æ–º –≥–æ—Ä–æ–¥–µ –≤—ã –∂–∏–≤–µ—Ç–µ?: ')
if age >= 12 and grade >= 7 and city == '–ú–æ—Å–∫–≤–∞':
    print('–î–æ—Å—Ç—É–ø —Ä–∞–∑—Ä–µ—à–µ–Ω.')
else:
    print('–î–æ—Å—Ç—É–ø –∑–∞–ø—Ä–µ—â–µ–Ω.')
```

 # –û–ø–µ—Ä–∞—Ç–æ—Ä or

```python
# –¥–ª—è –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –±–ª–æ–∫–∞ –∫–æ–¥–∞ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ–≥–æ –∏–∑ —É—Å–ª–æ–≤–∏–π.
```

```python
# –ú—ã –º–æ–∂–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –æ–±–∞ –ª–æ–≥–∏—á–µ—Å–∫–∏—Ö –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ:
# –¢–∞–∫–æ–π –∫–æ–¥ –ø—Ä–æ–≤–µ—Ä—è–µ—Ç, —á—Ç–æ –≤–æ–∑—Ä–∞—Å—Ç —É—á–µ–Ω–∏–∫–æ–≤ –æ—Ç –¥–≤–µ–Ω–∞–¥—Ü–∞—Ç–∏ –ª–µ—Ç –∏ —É—á–∞—Ç—Å—è
# –æ–Ω–∏ –ø–æ –∫—Ä–∞–π–Ω–µ–π –º–µ—Ä–µ –≤ 7 –∫–ª–∞—Å—Å–µ –∏ –∂–∏–≤—É—Ç –≤ –ú–æ—Å–∫–≤–µ –∏–ª–∏ –°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥–µ.
```

```python
age = int(input('–°–∫–æ–ª—å–∫–æ –≤–∞–º –ª–µ—Ç?: '))
grade = int(input('–í –∫–∞–∫–æ–º –∫–ª–∞—Å—Å–µ –≤—ã —É—á–∏—Ç–µ—Å—å?: '))
city = input('–í –∫–∞–∫–æ–º –≥–æ—Ä–æ–¥–µ –≤—ã –∂–∏–≤–µ—Ç–µ?: ')
if age >= 12 and grade >= 7 and (city == '–ú–æ—Å–∫–≤–∞' or city == '–°–∞–Ω–∫—Ç-–ü–µ—Ç–µ—Ä–±—É—Ä–≥'):
    print('–î–æ—Å—Ç—É–ø —Ä–∞–∑—Ä–µ—à–µ–Ω.')
else:
    print('–î–æ—Å—Ç—É–ø –∑–∞–ø—Ä–µ—â–µ–Ω.')
```

# –û–ø–µ—Ä–∞—Ç–æ—Ä not

```python
age = int(input('–°–∫–æ–ª—å–∫–æ –≤–∞–º –ª–µ—Ç?: '))
if not (age < 12):
    print('–î–æ—Å—Ç—É–ø —Ä–∞–∑—Ä–µ—à–µ–Ω.')
else:
    print('–î–æ—Å—Ç—É–ø –∑–∞–ø—Ä–µ—â–µ–Ω.')
```

# –ø–æ–ª–Ω–æ—Å—Ç—å—é —ç–∫–≤–∏–≤–∞–ª–µ–Ω—Ç–µ–Ω –∫–æ–¥—É –Ω–∏–∂–µ:

```python
age = int(input('–°–∫–æ–ª—å–∫–æ –≤–∞–º –ª–µ—Ç?: '))
if age >= 12:
    print('–î–æ—Å—Ç—É–ø —Ä–∞–∑—Ä–µ—à–µ–Ω.')
else:
    print('–î–æ—Å—Ç—É–ø –∑–∞–ø—Ä–µ—â–µ–Ω.')
```

# –ü—Ä–∏–æ—Ä–∏—Ç–µ—Ç—ã –ª–æ–≥–∏—á–µ—Å–∫–∏—Ö –æ–ø–µ—Ä–∞—Ç–æ—Ä–æ–≤

## –¥–∞–ª–µ–µ –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –ª–æ–≥–∏—á–µ—Å–∫–æ–µ —Å–ª–æ–∂–µ–Ω–∏–µ or.

# –î–ª—è —è–≤–Ω–æ–≥–æ —É–∫–∞–∑–∞–Ω–∏—è –ø–æ—Ä—è–¥–∫–∞ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —É—Å–ª–æ–≤–Ω—ã—Ö –æ–ø–µ—Ä–∞—Ç–æ—Ä–æ–≤ –∏—Å–ø–æ–ª—å–∑—É—é—Ç —Å–∫–æ–±–∫–∏

# –û–±–∞ –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞, and –∏ or, –≤—ã—á–∏—Å–ª—è—é—Ç—Å—è –ø–æ —É–∫–æ—Ä–æ—á–µ–Ω–Ω–æ–π —Å—Ö–µ–º–µ.

```python
5 > 100 and 10 > 0
# –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ 5 > 100. –û–Ω–æ –ª–æ–∂–Ω–æ (–ø–æ—Ç–æ–º—É —á—Ç–æ 5 –Ω–µ –º–æ–∂–µ—Ç
# –±—ã—Ç—å –±–æ–ª—å—à–µ 100).
```

```python
# –ê–Ω–∞–ª–æ–≥–∏—á–Ω–æ —Ä–∞–±–æ—Ç–∞–µ—Ç –æ–ø–µ—Ä–∞—Ç–æ—Ä or
# –ï—Å–ª–∏ —É—Å–ª–æ–≤–∏–µ —Å–ª–µ–≤–∞ –æ—Ç –æ–ø–µ—Ä–∞—Ç–æ—Ä–∞ or –∏—Å—Ç–∏–Ω–Ω–æ–µ, —Ç–æ —É—Å–ª–æ–≤–∏–µ —Å–ø—Ä–∞–≤–∞ –æ—Ç –Ω–µ–≥–æ
# –Ω–µ –ø—Ä–æ–≤–µ—Ä—è–µ—Ç—Å—è. –î–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ, —Ä–µ–∑—É–ª—å—Ç–∞—Ç –±—É–¥–µ—Ç –≥–∞—Ä–∞–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ
# –∏—Å—Ç–∏–Ω–Ω—ã–º –∏ –ø—Ä–æ–≤–µ—Ä–∫–∞ –æ—Å—Ç–∞–≤—à–µ–≥–æ—Å—è —É—Å–ª–æ–≤–∏—è —Å—Ç–∞–Ω–µ—Ç –ø—É—Å—Ç–æ–π —Ç—Ä–∞—Ç–æ–π
# –ø—Ä–æ—Ü–µ—Å—Å–æ—Ä–Ω–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏.
```

```python
10 > 0 or 5 > 100
# –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è —Ç–æ–ª—å–∫–æ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ 10 > 0. –û–Ω–æ –ø—Ä–∞–≤–¥–≤–∏–≤–æ, –∑–Ω–∞—á–∏—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç
# —Ç–æ–∂–µ –ø—Ä–∞–≤–¥–∏–≤, —Ç.–∫. –Ω–∞–º –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –æ–¥–Ω–æ–≥–æ –ø—Ä–∞–≤–¥–∏–≤–æ–≥–æ –≤—ã—Ä–∞–∂–µ–Ω–∏—è –ø—Ä–∏
# –æ–ø–µ—Ä–∞—Ç–æ—Ä–µ or.
```

```python
# Test1
True and False	# ==> FALSE

True and True	  # ==> TRUE

False and True	# ==> FALSE

False and False	# ==> FALSE

True or False	  # ==> TRUE

True or True	  # ==> TRUE

False or True	  # ==> TRUE

False or False	# ==> FALSE

not True	      # ==> FALSE

not False       # ==> TRUE
```

```python
# Test2
# –ö–∞–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –±—É–¥–µ—Ç –≤—ã–≤–µ–¥–µ–Ω–æ –Ω–∞ —ç–∫—Ä–∞–Ω –ø–æ—Å–ª–µ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è —Å–ª–µ–¥—É—é—â–µ–π
# –ø—Ä–æ–≥—Ä–∞–º–º—ã, –µ—Å–ª–∏ —Å –∫–ª–∞–≤–∏–∞—Ç—É—Ä—ã –≤–≤–µ–¥–µ–Ω–æ —á–∏—Å–ª–æ 7?
a = int(input())
if a >= 2 and a <= 17:
    b = 3
    p = a * a + b * b
else:
    b = 5
p = (a + b) * (a + b)
print(p)
```

x.

–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

# Test4 –ü—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç—å 2
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ x –∏ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç, –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç –ª–∏ –¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ —É–∫–∞–∑–∞–Ω–Ω—ã–º –ø—Ä–æ–º–µ–∂—É—Ç–∫–∞–º.
–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ x.
–ü—Ä–∏–º–µ—á–∞–Ω–∏–µ. –ï—Å–ª–∏ —Ç–æ—á–∫–∞ –≤—ã–∫–æ–ª–æ—Ç–∞—è, —Ç–æ –≥—Ä–∞–Ω–∏—Ü–∞ –Ω–µ –≤–∫–ª—é—á–∞–µ—Ç—Å—è, –µ—Å–ª–∏ —Ç–æ—á–∫–∞ –∑–∞–∫—Ä–∞—à–µ–Ω–Ω–∞—è, —Ç–æ –≥—Ä–∞–Ω–∏—Ü–∞ –≤–∫–ª—é—á–∞–µ—Ç—Å—è.

–¢–µ—Å—Ç–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ üü¢
Sample Input 1:

-44
Sample Output 1:

–ü—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç

```python
x = int(input())
if x <= -3 or x >= 7:
    print('–ü—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç')
else:
    print('–ù–µ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç')
```

# Test5 –ü—Ä–∏–Ω–∞–¥–ª–µ–∂–Ω–æ—Å—Ç—å 3
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –∫–æ—Ç–æ—Ä–∞—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ x –∏ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç, –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç –ª–∏ –¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ —É–∫–∞–∑–∞–Ω–Ω—ã–º –ø—Ä–æ–º–µ–∂—É—Ç–∫–∞–º.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ x.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.


```python
x = int(input())
if -30 < x <= -2 or 7 < x <= 25:
    print('–ü—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç')
else:
    print('–ù–µ –ø—Ä–∏–Ω–∞–¥–ª–µ–∂–∏—Ç')
```

# Test6 –ö—Ä–∞—Å–∏–≤–æ–µ —á–∏—Å–ª–æ üå∂Ô∏è
–ù–∞–∑–æ–≤–µ–º —á–∏—Å–ª–æ –∫—Ä–∞—Å–∏–≤—ã–º, –µ—Å–ª–∏ –æ–Ω–æ —è–≤–ª—è–µ—Ç—Å—è —á–µ—Ç—ã—Ä–µ—Ö–∑–Ω–∞—á–Ω—ã–º –∏ –¥–µ–ª–∏—Ç—Å—è –Ω–∞—Ü–µ–ª–æ –Ω–∞
7 –∏–ª–∏ –Ω–∞ 17. –ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É, –æ–ø—Ä–µ–¥–µ–ª—è—é—â—É—é, —è–≤–ª—è–µ—Ç—Å—è –ª–∏ –≤–≤–µ–¥—ë–Ω–Ω–æ–µ —á–∏—Å–ª–æ –∫—Ä–∞—Å–∏–≤—ã–º. –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ ¬´YES¬ª, –µ—Å–ª–∏ —á–∏—Å–ª–æ —è–≤–ª—è–µ—Ç—Å—è –∫—Ä–∞—Å–∏–≤—ã–º, –∏–ª–∏ ¬´NO¬ª –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.

–§–æ—Ä–º–∞—Ç –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ù–∞ –≤—Ö–æ–¥ –ø—Ä–æ–≥—Ä–∞–º–º–µ –ø–æ–¥–∞—ë—Ç—Å—è –Ω–∞—Ç—É—Ä–∞–ª—å–Ω–æ–µ —á–∏—Å–ª–æ.

–§–æ—Ä–º–∞—Ç –≤—ã—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
–ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ —Ç–µ–∫—Å—Ç –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É—Å–ª–æ–≤–∏–µ–º –∑–∞–¥–∞—á–∏.

–¢–µ—Å—Ç–æ–≤—ã–µ –¥–∞–Ω–Ω—ã–µ üü¢
Sample Input 1:

1043
Sample Output 1:

YES
Sample Input 2:

1045
Sample Output 2:

(7;2) ‚Äî —á–µ—Ä–Ω—ã–π –∫–∞–∫ –∏ (7;6). –ß—Ç–æ —ç—Ç–æ –Ω–∞–º –¥–∞–µ—Ç? –ê —ç—Ç–æ –Ω–∞–º –¥–∞–µ—Ç –æ—Ç–ª–∏—á–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ.

–ï–°–õ–ò —Å—É–º–º–∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–æ–≤ –ø–µ—Ä–≤–æ–π –∫–ª–µ—Ç–∫–∏ —á–µ—Ç–Ω–∞—è –ò —Å—É–º–º–∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–æ–≤ –≤—Ç–æ—Ä–æ–π –∫–ª–µ—Ç–∫–∏ —á–µ—Ç–Ω–∞—è –ò–õ–ò —Å—É–º–º–∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–æ–≤ –ø–µ—Ä–≤–æ–π –∫–ª–µ—Ç–∫–∏ –Ω–µ—á–µ—Ç–Ω–∞—è –ò —Å—É–º–º–∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–æ–≤ –≤—Ç–æ—Ä–æ–π –∫–ª–µ—Ç–∫–∏ –Ω–µ—á–µ—Ç–Ω–∞—è
–≤—ã–≤–æ–¥–∏–º –Ω–∞ —ç–∫—Ä–∞–Ω YES ‚Äî —Ç.–µ. –¥–≤–µ –∫–ª–µ—Ç–∫–∏ –æ–¥–Ω–æ–≥–æ —Ü–≤–µ—Ç–∞.
–ò–ù–ê–ß–ï —Ç.–µ. –µ—Å–ª–∏ –¥–∞–Ω–Ω—ã–µ –Ω–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç –≤—ã—à–µ–ø–µ—Ä–µ—á–∏—Å–ª–µ–Ω–Ω–æ–º—É —É—Å–ª–æ–≤–∏—é –≤—ã–≤–æ–¥–∏–º –Ω–∞ —ç–∫—Ä–∞–Ω NO.

```python
x1 = int (input())
y1 = int (input())
x2 = int (input())
y2 = int (input())
if (x1 + y1)%2 == 0 and (x2 + y2)%2 == 0 or (x1 + y1)%2 == 1 and (x2 + y2)%2 == 1:
print(‚ÄòYES‚Äô)
else:
print(‚ÄòNO‚Äô)
```

```python
# –ú–æ–∂–Ω–æ –ø–æ—Å—Ç—É–ø–∏—Ç—å –µ—â–µ –ø—Ä–æ—â–µ, –ø—Ä–æ—Å—Ç–æ —Å—Ä–∞–≤–Ω–∏—Ç—å –æ—Å—Ç–∞—Ç–∫–∏ –æ—Ç –¥–µ–ª–µ–Ω–∏—è –Ω–∞ 2 —Å—É–º–º
# –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–æ–≤ –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π –∫–ª–µ—Ç–∫–∏, –µ—Å–ª–∏ –æ–Ω–∏ —Ä–∞–≤–Ω—ã, —Ç–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ –∫–ª–µ—Ç–∫–∏
# —Ç–æ–∂–µ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ, –≤–µ–¥—å –µ—Å–ª–∏ –æ–±–µ –∫–ª–µ—Ç–∫–∏ —Å —á–µ—Ç–Ω—ã–º–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º–∏ —Ç–æ –æ—Å—Ç–∞—Ç–∫–∏ –±—É–¥—É—Ç
# –Ω—É–ª—è–º–∏, –∞ –µ—Å–ª–∏ –Ω–µ—á–µ—Ç–Ω—ã–µ —Ç–æ –µ–¥–∏–Ω–∏—Ü–µ–π:
x1 = int(input())
y1 = int(input())
x2 = int(input())
y2 = int(input())
if (x1 + y1)%2 == (x2 + y2)%2:
    print('YES')
else:
    print('NO')
```

```python
# –ï—Å–ª–∏ –∏ –¥–∞–ª—å—à–µ —Ä–∞—Å—Å—É–∂–¥–∞—Ç—å –ª–æ–≥–∏—á–µ—Å–∫–∏ –º–æ–∂–Ω–æ –ø—Ä–∏–π—Ç–∏ –∫ —Å–ª–µ–¥—É—é—â–µ–º—É.
# –°—É–º–º–∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–æ–≤ –∫–∞–∂–¥–æ–π –∏–∑ –∫–ª–µ—Ç–æ–∫ –º–æ–∂–µ—Ç –±—ã—Ç—å –ª–∏–±–æ —á–µ—Ç–Ω–æ–π –ª–∏–±–æ –Ω–µ—á–µ—Ç–Ω–æ–π.
# –ê —Å—É–º–º–∞ –¥–≤—É—Ö —á–µ—Ç–Ω—ã—Ö –∏–ª–∏ –¥–≤—É—Ö –Ω–µ—á–µ—Ç–Ω—ã—Ö —á–∏—Å–µ–ª –≤—Å–µ–≥–¥–∞ –±—É–¥–µ—Ç —á–µ—Ç–Ω–æ–µ —á–∏—Å–ª–æ. –û—Ç—Å—é–¥–∞
# –º—ã –º–æ–∂–µ–º –Ω–∞–ø–∏—Å–∞—Ç—å —Å–ª–µ–¥—É—é—â–µ–µ —É—Å–ª–æ–≤–∏–µ. –ï—Å–ª–∏ —Å—É–º–º–∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–æ–≤ –æ–±–µ–∏—Ö –∫–ª–µ—Ç–æ–∫
# —á–µ—Ç–Ω–æ–µ —á–∏—Å–ª–æ —Ç–æ –∑–Ω–∞—á–∏—Ç –∏ –∫–ª–µ—Ç–∫–∏ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ.
x1 = int(input())
y1 = int(input())
x2 = int(input())
y2 = int(input())
if (x1 + y1 + x2 + y2) % 2 == 0:
    print('YES')
else:
    print('NO')
```

## –®–∞—Ö–º–∞—Ç–Ω—ã–µ –∑–∞–¥–∞—á–∫–∏. –ö–∞–∫ —Ö–æ–¥—è—Ç —à–∞—Ö–º–∞—Ç–Ω—ã–µ —Ñ–∏–≥—É—Ä—ã

### –ó–∞–¥–∞—á–∞ –æ—á–µ–Ω—å –ø—Ä–æ—Å—Ç–∞—è, —á—Ç–æ–±—ã —Ä–µ—à–∏—Ç—å –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–∏—Ç—å —Å–µ–±–µ –ª–∞–¥—å—é –Ω–∞ —à–∞—Ö–º–∞—Ç–Ω–æ–º –ø–æ–ª–µ –∏ –ø—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –µ–µ —Ö–æ–¥. –õ–∞–¥—å—è —Ö–æ–¥–∏—Ç —Ç–æ–ª—å–∫–æ –ª–∏–±–æ –ø–æ –≤–µ—Ä—Ç–∏–∫–∞–ª–µ –≤–≤–µ—Ä—Ö –∏–ª–∏ –≤–Ω–∏–∑, –ª–∏–±–æ –ø–æ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª–∏ –≤–ª–µ–≤–æ –∏–ª–∏ –≤–ø—Ä–∞–≤–æ. –°—Ç–∞–Ω–æ–≤–∏—Ç—Å—è —è—Å–Ω–æ, —á—Ç–æ –æ–¥–Ω–∞ –∏–∑ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –∫–ª–µ—Ç–∫–∏ –≤—Å–µ–≥–¥–∞ –æ—Å—Ç–∞–µ—Ç—Å—è –Ω–µ–∏–∑–º–µ–Ω–Ω–æ–π, —Ç.–µ. –µ—Å–ª–∏ –ª–∞–¥—å—è —Ö–æ–¥–∏—Ç –ø–æ –≤–µ—Ä—Ç–∏–∫–∞–ª–µ –Ω–µ–∏–∑–º–µ–Ω–Ω–∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞ –• –µ—Å–ª–∏ –ø–æ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª–∏ —Ç–æ Y. –û—Ç—Å—é–¥–∞ –Ω–∞–ø–∏—à–µ–º —É—Å–ª–æ–≤–∏–µ –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º –±—É–¥–µ–º —Å—Ä–∞–≤–Ω–∏–≤–∞—Ç—å –µ—Å–ª–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã X –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π –∫–ª–µ—Ç–∫–∏ –æ–¥–∏–Ω–∞–∫–æ–≤—ã –∏–ª–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã Y –ø–µ—Ä–≤–æ–π –∏ –≤—Ç–æ—Ä–æ–π –∫–ª–µ—Ç–∫–∏ –æ–¥–∏–Ω–∞–∫–æ–≤—ã –∑–Ω–∞—á–∏—Ç –≤—ã–≤–æ–¥–∏–º YES ‚Äì –ª–∞–¥—å—è –º–æ–∂–µ—Ç –ø–æ–ø–∞—Å—Ç—å —Å –ø–µ—Ä–≤–æ–π –∫–ª–µ—Ç–∫–∏ –Ω–∞ –≤—Ç–æ—Ä—É—é, –∏–Ω–∞—á–µ ‚Äì NO, –Ω–µ –º–æ–∂–µ—Ç.

```python
x1 = int(input())
y1 = int(input())
x2 = int(input())
y2 = int(input())
if x1 == x2 or y1 == y2:
    print('YES')
else:
    print('NO')
```

### –•–æ–¥ –∫–æ—Ä–æ–ª—è
### –®–∞—Ö–º–∞—Ç–Ω—ã–π –∫–æ—Ä–æ–ª—å —Ö–æ–¥–∏—Ç –ø–æ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª–∏, –≤–µ—Ä—Ç–∏–∫–∞–ª–∏ –∏ –¥–∏–∞–≥–æ–Ω–∞–ª–∏, –Ω–æ —Ç–æ–ª—å–∫–æ –Ω–∞ 1 –∫–ª–µ—Ç–∫—É. –î–∞–Ω—ã –¥–≤–µ —Ä–∞–∑–ª–∏—á–Ω—ã–µ –∫–ª–µ—Ç–∫–∏ —à–∞—Ö–º–∞—Ç–Ω–æ–π –¥–æ—Å–∫–∏, –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, –º–æ–∂–µ—Ç –ª–∏ –∫–æ—Ä–æ–ª—å –ø–æ–ø–∞—Å—Ç—å —Å –ø–µ—Ä–≤–æ–π –∫–ª–µ—Ç–∫–∏ –Ω–∞ –≤—Ç–æ—Ä—É—é –æ–¥–Ω–∏–º —Ö–æ–¥–æ–º. –ü—Ä–æ–≥—Ä–∞–º–º–∞ –ø–æ–ª—É—á–∞–µ—Ç –Ω–∞ –≤—Ö–æ–¥ —á–µ—Ç—ã—Ä–µ —á–∏—Å–ª–∞ –æ—Ç 1 –¥–æ 8 –∫–∞–∂–¥–æ–µ, –∑–∞–¥–∞—é—â–∏–µ –Ω–æ–º–µ—Ä —Å—Ç–æ–ª–±—Ü–∞ –∏ –Ω–æ–º–µ—Ä —Å—Ç—Ä–æ–∫–∏ —Å–Ω–∞—á–∞–ª–∞ –¥–ª—è –ø–µ—Ä–≤–æ–π –∫–ª–µ—Ç–∫–∏, –ø–æ—Ç–æ–º –¥–ª—è –≤—Ç–æ—Ä–æ–π –∫–ª–µ—Ç–∫–∏. –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–µ—Å—Ç–∏ YES, –µ—Å–ª–∏ –∏–∑ –ø–µ—Ä–≤–æ–π –∫–ª–µ—Ç–∫–∏ —Ö–æ–¥–æ–º –∫–æ—Ä–æ–ª—è –º–æ–∂–Ω–æ –ø–æ–ø–∞—Å—Ç—å –≤–æ –≤—Ç–æ—Ä—É—é –∏–ª–∏ NO –≤ –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ.


### –ó–∞–¥–∞—á–∞ –ø–æ—Å–ª–æ–∂–Ω–µ–µ. –ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º —Ö–æ–¥. –ö–æ—Ä–æ–ª—å –º–æ–∂–µ—Ç —Ö–æ–¥–∏—Ç—å –≤–≤–µ—Ä—Ö –≤–Ω–∏–∑, –≤–ø—Ä–∞–≤–æ –≤–ª–µ–≤–æ –∏ –ø–æ –¥–∏–∞–≥–æ–Ω–∞–ª–∏ –Ω–æ —Ç–æ–ª—å–∫–æ –Ω–∞ –æ–¥–Ω—É –∫–ª–µ—Ç–∫—É. –¢.–µ. –µ—Å–ª–∏ —Ä–∞–∑–Ω–æ—Å—Ç—å –º–µ–∂–¥—É –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞–º–∏ X –∏ Y –±—É–¥–µ—Ç —Å–æ—Å—Ç–∞–≤–ª—è—Ç—å 1 –∏–ª–∏ -1 (–≤ —Å–ª—É—á–∞–µ –µ—Å–ª–∏ –ø–µ—Ä–µ—Ö–æ–¥–∏—Ç —Å –º–µ–Ω—å—à–µ–π –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç—ã –Ω–∞ –±–æ–ª—å—à—É—é) –∏–ª–∏ —Ä–∞–∑–Ω–æ—Å—Ç—å –æ–¥–Ω–æ–π –∏–∑ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç —Ä–∞–≤–Ω–∞ 0 —Ç–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ –∫–æ—Ä–æ–ª—å –º–æ–∂–µ—Ç –ø–µ—Ä–µ–π—Ç–∏ —Å –ø–µ—Ä–≤–æ–π –∫–ª–µ—Ç–∫–∏ –Ω–∞ –≤—Ç–æ—Ä—É—é.

```python
x1 = int(input())
y1 = int(input())
x2 = int(input())
y2 = int(input())

if (x1-x2 == 1 or x1-x2  == -1 or x1-x2  == 0) and (y1-y2 == 1 or y1-y2 == -1 or y1-y2 == 0):
    print ('YES')
else:
    print ('NO')
```

# –ß—Ç–æ –Ω–µ —Ç–∞–∫ —Å –∫–æ–¥–æ–º? –í—Å–µ —ç—Ç–æ –º–æ–∂–Ω–æ –±—ã–ª–æ –∑–∞–ø–∏—Å–∞—Ç—å –≥–æ—Ä–∞–∑–¥–æ –ª–∞–∫–æ–Ω–∏—á–Ω–µ–µ.

```python
x1 = int(input())
y1 = int(input())
x2 = int(input())
y2 = int(input())

if -1 <= x1-x2 <= 1 and -1 <= y1-y2 <= 1:
    print ('YES')
else:
    print ('NO')
```

# –ª–∏–±–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é –∞–±—Å–æ–ª—é—Ç–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã (–º–æ–¥—É–ª—è) —á–∏—Å–ª–∞ ‚Äì abs()

```python
x1 = int(input())
y1 = int(input())
x2 = int(input())
y2 = int(input())

if abs(x1-x2) <= 1 and abs(y1-y2) <= 1:
    print ('YES')
else:
    print ('NO')
```

### –•–æ–¥ —Å–ª–æ–Ω–∞
### –®–∞—Ö–º–∞—Ç–Ω—ã–π —Å–ª–æ–Ω —Ö–æ–¥–∏—Ç –ø–æ –¥–∏–∞–≥–æ–Ω–∞–ª–∏. –î–∞–Ω—ã –¥–≤–µ —Ä–∞–∑–ª–∏—á–Ω—ã–µ –∫–ª–µ—Ç–∫–∏ —à–∞—Ö–º–∞—Ç–Ω–æ–π –¥–æ—Å–∫–∏, –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, –º–æ–∂–µ—Ç –ª–∏ —Å–ª–æ–Ω –ø–æ–ø–∞—Å—Ç—å —Å –ø–µ—Ä–≤–æ–π –∫–ª–µ—Ç–∫–∏ –Ω–∞ –≤—Ç–æ—Ä—É—é –æ–¥–Ω–∏–º —Ö–æ–¥–æ–º.

### –ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º —Ö–æ–¥. –î–≤–∏–≥–∞—è —Å–ª–æ–Ω–∞ –ø–æ —à–∞—Ö–º–∞—Ç–Ω–æ–π –∫–ª–µ—Ç–∫–∏ –º–æ–∂–Ω–æ –∑–∞–º–µ—Ç–∏—Ç—å, —á—Ç–æ —Å–ª–æ–Ω –≤—Å–µ–≥–¥–∞ —Ö–æ–¥–∏—Ç –ø–æ –¥–∏–∞–≥–æ–Ω–∞–ª—è–º –∫–≤–∞–¥—Ä–∞—Ç–∞, —Ç.–µ. –µ—Å–ª–∏ –ø–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–µ X –æ–Ω –ø–µ—Ä–µ–¥–≤–∏–Ω–µ—Ç—Å—è –Ω–∞ 5 –∫–ª–µ—Ç–æ–∫ —Ç–æ –∏ –ø–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–µ Y –æ–Ω –ø–µ—Ä–µ–¥–≤–∏–Ω–µ—Ç—Å—è –Ω–∞ 5 –∫–ª–µ—Ç–æ–∫. –û—Ç—Å—é–¥–∞ –º–æ–∂–µ–º —Å–¥–µ–ª–∞—Ç—å –≤—ã–≤–æ–¥, —á—Ç–æ –º–æ–¥—É–ª—å —Ä–∞–∑–Ω–æ—Å—Ç–∏ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–æ–≤ X1 –∏ X2 –∏ Y1 –∏ Y2 –≤—Å–µ–≥–¥–∞ –±—É–¥–µ—Ç —Ä–∞–≤–µ–Ω

```python
x1 = int(input())
y1 = int(input())
x2 = int(input())
y2 = int(input())
if abs(x1 - x2) == abs(y1 - y2):
    print('YES')
else:
    print('NO')
```

### –•–æ–¥ —Ñ–µ—Ä–∑—è
### –®–∞—Ö–º–∞—Ç–Ω—ã–π —Ñ–µ—Ä–∑—å —Ö–æ–¥–∏—Ç –ø–æ –¥–∏–∞–≥–æ–Ω–∞–ª–∏, –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª–∏ –∏–ª–∏ –≤–µ—Ä—Ç–∏–∫–∞–ª–∏. –î–∞–Ω—ã –¥–≤–µ —Ä–∞–∑–ª–∏—á–Ω—ã–µ –∫–ª–µ—Ç–∫–∏ —à–∞—Ö–º–∞—Ç–Ω–æ–π –¥–æ—Å–∫–∏, –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, –º–æ–∂–µ—Ç –ª–∏ —Ñ–µ—Ä–∑—å –ø–æ–ø–∞—Å—Ç—å —Å –ø–µ—Ä–≤–æ–π –∫–ª–µ—Ç–∫–∏ –Ω–∞ –≤—Ç–æ—Ä—É—é –æ–¥–Ω–∏–º —Ö–æ–¥–æ–º.

### –ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º —Ö–æ–¥ —Ñ–µ—Ä–∑—è. –≠—Ç–∞ —Ñ–∏–≥—É—Ä–∞ —Ö–æ–¥–∏—Ç –∫–∞–∫ –∫–æ—Ä–æ–ª—å, –Ω–æ —É–∂–µ –Ω–∞ –ª—é–±–æ–µ –¥–æ—Å—Ç—É–ø–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–ª–µ—Ç–æ–∫, –Ω—É –∏–ª–∏ –º–æ–∂–µ–º —Å–∫–∞–∑–∞—Ç—å, —á—Ç–æ —Ñ–µ—Ä–∑—å —Ö–æ–¥–∏—Ç –∏ –∫–∞–∫ –ª–∞–¥—å—è –∏ –∫–∞–∫ —Å–ª–æ–Ω, –º—ã —É–∂–µ –∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞–ª–∏ –∏ –ø–∏—Å–∞–ª–∏ –∫–æ–¥ –¥–ª—è —ç—Ç–∏—Ö —Ñ–∏–≥—É—Ä –ø–æ—ç—Ç–æ–º—É –ø—Ä–æ—Å—Ç–æ –æ–±—ä–µ–¥–∏–Ω–∏–º –¥–≤–∞ —É—Å–ª–æ–≤–∏—è –≤ –æ–¥–Ω–æ:

```python
x1 = int(input())
y1 = int(input())
x2 = int(input())
y2 = int(input())

if abs(x1-x2) <= 1 and abs(y1-y2) <= 1 or x1 == x2 or y1 == y2:
    print ('YES')
else:
    print ('NO')
```

### –•–æ–¥ –∫–æ–Ω—è
### –®–∞—Ö–º–∞—Ç–Ω—ã–π –∫–æ–Ω—å —Ö–æ–¥–∏—Ç –±—É–∫–≤–æ–π ‚Äú–ì‚Äù ‚Äî –Ω–∞ –¥–≤–µ –∫–ª–µ—Ç–∫–∏ –ø–æ –≤–µ—Ä—Ç–∏–∫–∞–ª–∏ –≤ –ª—é–±–æ–º –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏–∏ –∏ –Ω–∞ –æ–¥–Ω—É –∫–ª–µ—Ç–∫—É –ø–æ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª–∏, –∏–ª–∏ –Ω–∞–æ–±–æ—Ä–æ—Ç. –î–∞–Ω—ã –¥–≤–µ —Ä–∞–∑–ª–∏—á–Ω—ã–µ –∫–ª–µ—Ç–∫–∏ —à–∞—Ö–º–∞—Ç–Ω–æ–π –¥–æ—Å–∫–∏, –æ–ø—Ä–µ–¥–µ–ª–∏—Ç–µ, –º–æ–∂–µ—Ç –ª–∏ –∫–æ–Ω—å –ø–æ–ø–∞—Å—Ç—å —Å –ø–µ—Ä–≤–æ–π –∫–ª–µ—Ç–∫–∏ –Ω–∞ –≤—Ç–æ—Ä—É—é –æ–¥–Ω–∏–º —Ö–æ–¥–æ–º

### –°–∞–º–∞—è –∏–Ω—Ç–µ—Ä–µ—Å–Ω–∞—è —Ñ–∏–≥—É—Ä–∞. –ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞–≤ —Ö–æ–¥ –∫–æ–Ω—è –±—É–∫–≤–æ–π "–ì" –º–æ–∂–Ω–æ —É–≤–∏–¥–µ—Ç—å —á—Ç–æ –µ—Å–ª–∏ –∫–æ–Ω—å —Ö–æ–¥–∏—Ç –≤–Ω–∏–∑ –∏–ª–∏ –≤–≤–µ—Ä—Ö –±—É–∫–≤–æ–π —Ç–æ –µ–≥–æ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞ –ø–æ X –º–µ–Ω—è–µ—Ç—Å—è –Ω–∞ 1 –∞ –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞ –ø–æ Y –Ω–∞ 2, –µ—Å–ª–∏ –≤–ª–µ–≤–æ –∏ –≤–ø—Ä–∞–≤–æ —Ç–æ –Ω–∞–æ–±–æ—Ä–æ—Ç X –Ω–∞ 2 –∞ Y –Ω–∞ 1. –ò—Å—Ö–æ–¥—è –∏–∑ —ç—Ç–æ–≥–æ –º–æ–∂–Ω–æ –Ω–∞–ø–∏—Å–∞—Ç—å –∫–æ–¥, —á—Ç–æ –µ—Å–ª–∏ —Ä–∞–∑–Ω–æ—Å—Ç—å –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç X1 –∏ X2 —É–º–µ–Ω—å—à–∏–ª–∞—Å—å –∏–ª–∏ —É–≤–µ–ª–∏—á–∏–ª–∞—Å—å –Ω–∞ 1 –∏ –ø—Ä–∏ —ç—Ç–æ–º —Ä–∞–∑–Ω–æ—Å—Ç—å –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç Y1 –∏ Y2 —É–º–µ–Ω—å—à–∞–ª–∞—Å—å –∏–ª–∏ —É–≤–µ–ª–∏—á–∏–ª–∞—Å—å –Ω–∞ 2 –∏–ª–∏ –µ—Å–ª–∏ —Ä–∞–∑–Ω–æ—Å—Ç—å –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç X1 –∏ X2 —É–º–µ–Ω—å—à–∏–ª–∞—Å—å –∏–ª–∏ —É–≤–µ–ª–∏—á–∏–ª–∞—Å—å –Ω–∞ 2 –∏ –ø—Ä–∏ —ç—Ç–æ–º —Ä–∞–∑–Ω–æ—Å—Ç—å –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç Y1 –∏ Y2 —É–º–µ–Ω—å—à–∞–ª–∞—Å—å –∏–ª–∏ —É–≤–µ–ª–∏—á–∏–ª–∞—Å—å –Ω–∞ 1 —Ç–æ –≤—ã–≤–æ–¥–∏–º YES –∏–Ω–∞—á–µ NO

```python
x1 = int(input())
y1 = int(input())
x2 = int(input())
y2 = int(input())

if (x1 - 1 == x2 or x1 + 1 == x2) and (y1 - 2 == y2 or y1 + 2 == y2):
    print ('YES')
elif (x1 - 2 == x2 or x1 + 2 == x2) and (y1 - 1 == y2 or y1 + 1 == y2):
    print ('YES')
else:
    print ('NO')
```

### –û—á–µ–Ω—å —Å—Ç—Ä–∞—à–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ –Ω–∞ —Å–∞–º–æ–º –¥–µ–ª–µ, –Ω–æ —Ä–∞–±–æ—á–µ–µ. –ú–æ–∂–Ω–æ –∏ –Ω—É–∂–Ω–æ –±—ã–ª–æ –ø—Ä–∏–º–µ–Ω–∏—Ç—å –º–æ–¥—É–ª—å —Ä–∞–∑–Ω–∏—Ü—ã –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç –∏ –≤–≤–µ—Å—Ç–∏ –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã–µ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ.

```python
x1 = int(input())
y1 = int(input())
x2 = int(input())
y2 = int(input())
dx = abs(x1 - x2) # –º–æ–¥—É–ª—å —Ä–∞–∑–Ω–∏—Ü—ã –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç X
dy = abs(y1 - y2) # –º–æ–¥—É–ª—å —Ä–∞–∑–Ω–∏—Ü—ã –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç Y
if dx == 1 and dy == 2 or dx == 2 and dy == 1:
# –µ—Å–ª–∏ —Ä–∞–∑–Ω–æ—Å—Ç—å X1 –∏ X2 —Ä–∞–≤–Ω–∞ –æ–¥–Ω–æ–π –∫–ª–µ—Ç–∫–µ –∞ Y1 –∏ Y2 –¥–≤—É–º –∏–ª–∏ –Ω–∞–æ–±–æ—Ä–æ—Ç X1 –∏ Y1
# –¥–≤—É–º –∫–ª–µ–∫—Ç–∞–º –∞ Y1 –∏ Y2 –æ–¥–Ω–æ–π –∫–ª–µ—Ç–∫–µ —Ç–æ –≤—ã–≤–æ–¥–∏–º –Ω–∞ —ç–∫—Ä–∞–Ω YES
    print('YES')
else:
    print('NO')
```
